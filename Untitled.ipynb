```python
import scipy
import numpy as np
from scipy.special import zeta, gamma

import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
x = np.linspace(1,4,200)
y = zeta(x)
yy = zeta(x, q=5)

plt.figure(figsize=(10,5), dpi=196)
plt.plot(x,y)
plt.plot(x, yy)
```
